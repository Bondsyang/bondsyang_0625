```python
import random

play_count = 0 #route區，一定會執行

while(True):

    min = 1
    max = 100
    count = 0
    guess_value = random.randint(min,max)  #前面一定要比後面小
    print("==========猜數字遊戲===========\n")
    while(True):
        count += 1
        try:
            keyin = int(input(f"猜數字範圍{min}~{max}"))
                        
            if not (min <= keyin <= max):
                raise Exception(f"輸入範圍請在{min}~{max}")
            else:
                if(keyin ==  guess_value):   #連在一起加()；嚴格的放最前面
                    print(f"賓果!猜對了，答案是:{guess_value}")
                    print(f"您猜了{count}次")
                    break

                elif keyin > guess_value:
                    print("再小一點")
                    max = keyin - 1
                
                else:
                    print("再大一點")
                    min = keyin + 1
        
        except ValueError as e:
            print("格式錯誤",e)
        
        except Exception as e:
            print(e)
        
        
        print(f"您已經猜{count}次")
      
    play_count += 1
    retry = input("請問是否還要繼續嗎？(y,n)")
        
    if retry == "y":
        continue
    else:
        print(f"遊戲結束，感謝您遊玩此遊戲{play_count}次")
    break
```
