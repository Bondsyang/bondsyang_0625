```python
import numpy as np
np_array = np.array([1, 2, 3, 4]) #nd_array 必需要相同型別
np_array
```

```python
py_list = [1, 2, 3, 4]
new_list = []
for ele in py_list:
    print(ele)
    new_list.append(ele*10)
new_list
```

```python
py_list = [1, 2, 3, 4]
#使用綜合表達式 (comprehension)
[ele*10 for ele in py_list] #list, dict要配合for…in迴圈
```

```python
np_array = np.array([1, 2, 3, 4])
#numpy可以使用陣列運算(向量運算) -> for 大資料量運算
np_array *= 10
np_array
```

```python
#利用numpy和matplotlib產生正弦波和餘弦波
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="正弦波")
plt.plot(x, y2, label="餘弦波")
plt.title("正弦波與餘弦波")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
