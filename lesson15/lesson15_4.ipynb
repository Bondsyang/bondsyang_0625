```python
import numpy as np

scores = np.random.randint(50, 100, size=(50, 5))
scores
```

```python
import pandas as pd

df = pd.DataFrame(scores,
            index = [f"學生{i+1}" for i in range(50)],
            columns = ["國文", "英文", "數學", "地理", "歷史"]
            )  #建立DataFrame，index為學生編號，columns為科目名稱
```

```python
#透過欄的選取

df["國文"]  #取得國文成績
df["英文"]  #取得英文成績
df[["國文","英文"]]
```

```python
#透過index還取
df.loc["學生1"]
df.loc[["學生1", "學生2"]]
```

```python
#透過index索引值選取
df.iloc[0]
df.iloc[[0, 1]]
```

```python
df.iloc[:5] #取得前五個學生的成績
df.iloc[-5:] #取得最後五個學生的成績
```

```python
import numpy as np
total = np.sum(df,axis=1)
mean = np.mean(df,axis=1)
mean
```

```python
df["總分"] = total
df["平均"] = mean
df
```

```python
five_df = df.iloc[:5] 
five_df["總分"] >= 380  #可以用比較運算子
five_df[five_df["總分"] >= 380]  #取得總分大於等於380的學生資料，series顯示法
```
